```python
from frontend import App

# create an app
app = App.create("plastic")

# create a sphere tetrahedral mesh
r = 0.25
V, F, T = app.mesh.icosphere(r=r, subdiv_count=4).tetrahedralize()
app.asset.add.tet("sphere", V, F, T)

# create a square sheet mesh
V, F = app.mesh.square(res=32, ex=[1, 0, 0], ey=[0, 0, 1])
app.asset.add.tri("sheet", V, F)

# create a scene
scene = app.scene.create()

# add compressing squares
gap = 0.01
scene.add("sheet").at(0, 3 * r, 0).pin().move_by(
    [0, -3 * r + gap, 0], t_start=0, t_end=1
).move_by([0, 1, 0], t_start=1.5, t_end=2)
floor = scene.add("sheet").at(0, 0, 0)
floor.param.set("friction", 1.0)
floor.pin().move_by([0, 0, 0], t_start=0, t_end=1)

# add a squishy ball
ball = scene.add("sphere").at(0, r + gap, 0)
ball.param.set("plasticity", 0.5).set("plasticity-threshold", 0.5).set("friction", 1.0)

# compile the scene and report stats
scene = scene.build().report()

# preview the scene
scene.preview()
```

```python
# create a new session with the compiled scene
session = app.session.create(scene)

# set session parameters
session.param.set("frames", 180).set("dt", 0.01)

# build this session
session = session.build()

# start the simulation and live-preview the results
session.start().preview()

# also show simulation logs in realtime
session.stream()
```

```python
# create an animation from the simulation results
session.animate()
```
