```python
# import our frontend
from frontend import App

# make an app
app = App.create("intersect-1")

# create a square mesh resolution 64 spanning the xz plane
V, F = app.mesh.square(res=64, ex=[1, 0, 0], ey=[0, 0, 1])

# add to the asset and name it "sheet"
app.asset.add.tri("sheet", V, F)

# create an icosphere mesh radius 0.5
V, F = app.mesh.icosphere(r=0.5, subdiv_count=3)

# add to the asset and name it "sphere"
app.asset.add.tri("sphere", V, F)

# create a scene
scene = app.scene.create()

# add the sheet asset to the scene with an vertical offset
obj = scene.add("sheet").at(0, 0, 0)

# add a sphere mesh at a lower position and set it static collider
# this yields intersections
scene.add("sphere").at(0, -0.25, 0).pin()

# compile the scene and report stats
scene = scene.build().report()

# preview the initial scene
scene.preview()
```

```python
# create a new session with the compiled scene
session = app.session.create(scene)
```
